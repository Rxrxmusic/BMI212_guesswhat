### GENDER ANALYSIS ###

```python
import pandas as pd
```

```python
df_intern = pd.read_csv('exp4_intern.csv')
df_intern.head()
```

```python
df_ssbd = pd.read_csv('exp1_ssbd.csv')
df_ssbd.head()
```

```python
# Cathy's models
def extract_relevant_part(path):
    start = path.find('home/cathyhou/dataset/guesswhat/') + len('home/cathyhou/dataset/guesswhat/')
    end = len(path)
    return path[start:end]

df_intern['Path'] = df_intern['Path'].apply(extract_relevant_part)
df_ssbd['Path'] = df_ssbd['Path'].apply(extract_relevant_part)
```

```python
df_dino = pd.read_csv('test_statistics_multilayer_b16.csv')
df_dino.head()
```

```python
# Rachel's model
def extract_relevant_part(path):
    start = path.find('features_dinovitb/') + len('features_dinovitb/')
    end = path.find('/features.npy')
    return path[start:end]

df_dino['Path'] = df_dino['Path'].apply(extract_relevant_part)
df_dino['Path'] = df_dino['Path'] + '/GuessWhat.mp4'
```

```python
df_data = pd.read_csv('dataset_1_clean.csv')
df_data.rename(columns={'video_key': 'Path'}, inplace=True)
df_data.head()
```

```python
merged_intern = pd.merge(df_intern, df_data, on='Path')
merged_ssbd = pd.merge(df_ssbd, df_data, on='Path')
merged_dino = pd.merge(df_dino, df_data, on='Path')
```

```python
# sanity check
len(merged_intern.index) == len(df_intern.index)
len(merged_ssbd.index) == len(df_ssbd.index)
len(merged_dino.index) == len(df_dino.index)
```

```python
import numpy as np

# Functions to calculate metrics
def calculate_metrics(df):
    metrics = {}
    TP = np.sum((df['True Label'] == 1) & (df['Predicted Label'] == 1))
    TN = np.sum((df['True Label'] == 0) & (df['Predicted Label'] == 0))
    FP = np.sum((df['True Label'] == 0) & (df['Predicted Label'] == 1))
    FN = np.sum((df['True Label'] == 1) & (df['Predicted Label'] == 0))
    
    accuracy = (TP + TN) / len(df)
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    false_positive_rate = FP / (FP + TN) if (FP + TN) > 0 else 0
    false_negative_rate = FN / (FN + TP) if (FN + TP) > 0 else 0
    selection_rate = np.sum(df['Predicted Label'] == 1) / len(df)
    count = len(df)
    
    metrics['accuracy'] = accuracy
    metrics['precision'] = precision
    metrics['recall'] = recall
    metrics['f1_score'] = f1_score
    metrics['false_positive_rate'] = false_positive_rate
    metrics['false_negative_rate'] = false_negative_rate
    metrics['selection_rate'] = selection_rate
    metrics['count'] = count
    
    return pd.Series(metrics)

# Calculate metrics for each gender
metrics_by_gender_intern = merged_intern.groupby('gender').apply(calculate_metrics).reset_index()
metrics_by_gender_ssbd = merged_ssbd.groupby('gender').apply(calculate_metrics).reset_index()
metrics_by_gender_dino = merged_dino.groupby('gender').apply(calculate_metrics).reset_index()

# Add a model identifier to each DataFrame
metrics_by_gender_ssbd['Model'] = 'ssbd'
metrics_by_gender_intern['Model'] = 'intern'
metrics_by_gender_dino['Model'] = 'dinov2'

# Combine DataFrames
metrics_combined = pd.concat([metrics_by_gender_ssbd, metrics_by_gender_intern, metrics_by_gender_dino])

print(metrics_combined)
```

```python
metrics_combined.to_csv('results/models_by_gender.csv', index=False) 
```

```python
import matplotlib.pyplot as plt

# Plotting function for combined metrics
def plot_combined_metrics(metrics_df, metric_name):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Create a list of unique age buckets and models
    age_buckets = metrics_df['gender'].unique()
    models = metrics_df['Model'].unique()
    
    # Set the bar width and positions
    bar_width = 0.2
    bar_positions = np.arange(len(age_buckets))
    
    # Plot bars for each model
    for i, model in enumerate(models):
        model_df = metrics_df[metrics_df['Model'] == model]
        ax.bar(bar_positions + i * bar_width, model_df[metric_name], bar_width, label=model)
    
    # Set labels and titles
    ax.set_xlabel('Gender')
    ax.set_ylabel(metric_name)
    ax.set_title(f'{metric_name} by gender and model')
    ax.set_xticks(bar_positions + bar_width / 2)
    ax.set_xticklabels(age_buckets)
    ax.legend(title='Model')
    
    # Save the figure
    plt.savefig(f'results/{metric_name}_by_gender.png')
    plt.show()
    plt.clf()  # Clear the current figure to avoid overlap

# Plot all metrics and save the figures
metrics_list = ['accuracy', 'precision', 'recall', 'f1_score', 'false_positive_rate', 'false_negative_rate', 'selection_rate', 'count']

for metric in metrics_list:
    plot_combined_metrics(metrics_combined, metric)
```

### AGE ANALYSIS ###

```python
# Define age buckets
bins = [0, 3, 6, 9, float('inf')]
labels = ['0-3', '4-6', '7-9', '9+']
merged_ssbd['Age Bucket'] = pd.cut(merged_ssbd['age'], bins=bins, labels=labels, right=False)
merged_intern['Age Bucket'] = pd.cut(merged_intern['age'], bins=bins, labels=labels, right=False)
merged_dino['Age Bucket'] = pd.cut(merged_dino['age'], bins=bins, labels=labels, right=False)
```

```python
metrics_by_age_bucket_ssbd = merged_ssbd.groupby('Age Bucket').apply(calculate_metrics).reset_index()
metrics_by_age_bucket_intern = merged_intern.groupby('Age Bucket').apply(calculate_metrics).reset_index()
metrics_by_age_bucket_dino = merged_dino.groupby('Age Bucket').apply(calculate_metrics).reset_index()
```

```python
# Add a model identifier to each DataFrame
metrics_by_age_bucket_ssbd['Model'] = 'ssbd'
metrics_by_age_bucket_intern['Model'] = 'intern'
metrics_by_age_bucket_dino['Model'] = 'dinov2'

# Combine DataFrames
metrics_combined_age = pd.concat([metrics_by_age_bucket_ssbd, metrics_by_age_bucket_intern, metrics_by_age_bucket_dino])

print(metrics_combined_age)
```

```python
metrics_combined_age.to_csv('results/models_by_age.csv', index=False) 
```
